```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Kamal","Nimal","Sunil","Amal","Kasun"],
    "Math": [78,65,90,55,80],
    "Science": [82,70,88,60,85],
    "English": [75,68,92,58,79]
}

df = pd.DataFrame(data)

print(df)
print("Average Math:", df["Math"].mean())

df["Math"].plot(kind="bar")
plt.title("Math Scores")
plt.show()
```
